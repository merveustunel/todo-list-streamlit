```python
import threading
import time
import os
from pyngrok import ngrok

# Stop all ngrok processes before starting a new one
os.system("killall ngrok")

# Eğer eski tüneller açıksa kapatalım
ngrok.kill()

# Streamlit'i arka planda başlat
def run_app():
    os.system("streamlit run app.py --server.port 8501 --server.headless true")

thread = threading.Thread(target=run_app, daemon=True)
thread.start()

time.sleep(5)  # biraz bekle

# 8501 portunu ngrok'a bağla
public_url = ngrok.connect(8501, "http")
print("Uygulama linki:", public_url)
```
